```python
# a.
n = 40000
Z = np.random.randn(n)
plt.step(sorted(Z), np.arange(1, n+1)/float(n))
plt.xlabel('Random number from nornal distribution')
plt.ylabel('Cumulative Probability')
plt.xlim(-3,3)
```

```python
# b.
ks = [1,8,64,512]

for k in ks:
    Z = np.sum(np.sign(np.random.randn(n,k))*np.sqrt(1./k), axis=1)
    plt.step(sorted(Z), np.arange(1, n+1)/float(n), label=k)
    plt.xlabel('Random number from nornal distribution')
    plt.ylabel('Cumulative Probability')
    plt.xlim(-3,3)
    plt.legend()
```
